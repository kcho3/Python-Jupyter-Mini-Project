```python

#uploading 
import pandas as pd
from matplotlib import pyplot as plt

data = pd.read_csv('/selectedcountries.csv')

#accesssing data from columns
daycount = data['Days']
AF = data['Afghanistan']
AL = data['Albania']
AG = data['Algeria']

#plotting
plt.plot(daycount, AF, label="Afghanistan")
plt.plot(daycount, AL, label="Albania")
plt.plot(daycount, AG, label="Algeria")

#labelling
plt.legend()
plt.title('Confirmed Coronavirus Cases on Selected Countries over 100 Days')
plt.ylabel('Number of Confirmed Cases')
plt.xlabel('Number of Days')

plt.savefig("plot.png")
plt.show()
```
